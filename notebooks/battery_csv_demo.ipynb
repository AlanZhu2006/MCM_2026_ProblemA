# Battery CSV Demo (Minimal)This small notebook uses the path resolver to locate the trajectory CSV and plot SOC over time.

```python
from battery_csv_path_utils import resolve_csv_path
csv_path = resolve_csv_path()
print('CSV path resolved to:', csv_path)
```

```python
import pandas as pd
df = pd.read_csv(csv_path)
print('Loaded', len(df), 'rows')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,3))
plt.plot(df['time_s'], df['SOC'])
plt.xlabel('Time (s)')
plt.ylabel('SOC')
plt.title('SOC over Time')
plt.tight_layout()
plt.show()
```
